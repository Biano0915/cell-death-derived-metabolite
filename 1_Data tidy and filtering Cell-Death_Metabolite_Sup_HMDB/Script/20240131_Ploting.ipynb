### Tidy raw data

```python
# negative data
neg_measure.columns = list(neg_measure.iloc[1]) # set column names
neg_measure = neg_measure.drop([0,1]) # drop first two rows

# convert column to numeric, except for the first column
neg_measure.iloc[:,1:] = neg_measure.iloc[:,1:].apply(pd.to_numeric)

## Filter the columns containing specific text in their names
dmso_columns = [col for col in neg_measure.columns if 'DMSO' in col]
sts_columns = [col for col in neg_measure.columns if 'STS' in col]
fas_columns = [col for col in neg_measure.columns if 'Fas' in col]
rsl_columns = [col for col in neg_measure.columns if 'RSL' in col]
nec_columns = [col for col in neg_measure.columns if 'Nec' in col]
bk_columns = [col for col in neg_measure.columns if 'BK' in col][0]

# Calculate the mean of each group and add it to original data
neg_measure['DMSO mean'] = neg_measure[dmso_columns].mean(axis=1)
neg_measure['STS mean'] = neg_measure[sts_columns].mean(axis=1)
neg_measure['Fas mean'] = neg_measure[fas_columns].mean(axis=1)
neg_measure['RSL mean'] = neg_measure[rsl_columns].mean(axis=1)
neg_measure['Nec mean'] = neg_measure[nec_columns].mean(axis=1)

# Multiply the BK column by 10 and add it to original data
neg_measure['BK 10x'] = neg_measure[bk_columns] * 10

# remove the row with 0 identification
neg_measure_clean = neg_measure[neg_measure['Identifications'] != 0]

# remove the unnecessary columns
neg_measure_clean = neg_measure_clean.drop(columns=['Neutral mass (Da)', 'Chromatographic peak width (min)'])

# rename the group names
for i in dmso_columns:
    neg_measure_clean.rename(columns = {i: f'dmso_{dmso_columns.index(i) + 1}'}, inplace = True)

for i in sts_columns:
    neg_measure_clean.rename(columns = {i: f'sts_{sts_columns.index(i) + 1}'}, inplace = True)

for i in fas_columns:
    neg_measure_clean.rename(columns = {i: f'fas_{fas_columns.index(i) + 1}'}, inplace = True)

for i in rsl_columns:
    neg_measure_clean.rename(columns = {i: f'rsl_{rsl_columns.index(i) + 1}'}, inplace = True)

for i in nec_columns:
    neg_measure_clean.rename(columns = {i: f'nec_{nec_columns.index(i) + 1}'}, inplace = True)

# rename BK column
neg_measure_clean.rename(columns = {bk_columns: 'BK'}, inplace = True)

# change column to numeric data (start at column1)
neg_measure_clean.iloc[:,1:] = neg_measure_clean.iloc[:,1:].apply(pd.to_numeric)

neg_measure_clean
```

Generate score

```python
# function for assign rank
def assign_rank(row):
    count = sum(row < 0.05)
    if count == 1:
        return 2.5
    elif count == 2:
        return 5
    elif count == 3:
        return 7.5
    elif count == 4:
        return 10
    else:
        return 20
```

```python
# transfer to all numeric
neg_measure_rk = neg_measure_rk.apply(pd.to_numeric, errors='coerce')

neg_measure['Rank'] = neg_measure_rk.apply(assign_rank, axis=1)

# remove the row which compound didn't show up in the identification data
counts = neg_measure['Compound'].isin(neg_iden_merge['Compound']).value_counts()
print(f'* Removed {counts[False]}/{len(neg_measure)} compound didn\'t show up in the identification data, remain {counts[True]} compounds\n')
neg_measure = neg_measure[neg_measure['Compound'].isin(neg_iden_merge['Compound'])]

neg_measure.to_csv('/Users/biano/Desktop/20231204_Cell-Death_Metabolite_Sup_HMDB/20240131 Ploting/Processed/3_neg_measurement_p_rank.csv', index = False)
neg_measure
```
