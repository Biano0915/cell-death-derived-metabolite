```python
# import identification data
# the file path "/Users/biano/Desktop/" may be different depend on the folder location
neg_iden = pd.read_csv('/Users/biano/Desktop/20231204_Cell-Death_Metabolite_Sup_HMDB/Raw Data/NEG HMDB output/NEG HMDB identification.csv')
pos_iden = pd.read_csv('/Users/biano/Desktop/20231204_Cell-Death_Metabolite_Sup_HMDB/Raw Data/POS HMDB output/POS HMDB identification.csv')
```

```python
# import measurement data
neg_measure = pd.read_csv('/Users/biano/Desktop/20231204_Cell-Death_Metabolite_Sup_HMDB/Raw Data/NEG HMDB output/NEG HMDB measurement.csv', header = 2)
pos_measure = pd.read_csv('/Users/biano/Desktop/20231204_Cell-Death_Metabolite_Sup_HMDB/Raw Data/POS HMDB output/POS HMDB measurement.csv', header = 2)
```

```python
# Negative Measurement Data
## Filter the columns containing specific text in their names
dmso_columns = [col for col in neg_measure.columns if 'DMSO' in col]
sts_columns = [col for col in neg_measure.columns if 'STS' in col]
fas_columns = [col for col in neg_measure.columns if 'Fas' in col]
rsl_columns = [col for col in neg_measure.columns if 'RSL' in col]
nec_columns = [col for col in neg_measure.columns if 'Nec' in col]
bk_columns = [col for col in neg_measure.columns if 'BK' in col][0]

## Calculate the mean each group
dmso_mean = neg_measure[dmso_columns].mean(axis=1)
sts_mean = neg_measure[sts_columns].mean(axis=1)
fas_mean = neg_measure[fas_columns].mean(axis=1)
rsl_mean = neg_measure[rsl_columns].mean(axis=1)
nec_mean = neg_measure[nec_columns].mean(axis=1)

# Multiply the BK column by 10
bk_10x = neg_measure[bk_columns] * 10

# Get compound ID
Compound_id = neg_measure

## Create the new filtered dataframe
neg_filter = pd.DataFrame({'Compound': neg_measure['Compound'],
                           'iden': neg_measure['Identifications'],
                           'DMSO mean': dmso_mean,
                           'STS mean': sts_mean,
                           'Fas mean': fas_mean,
                           'RSL mean': rsl_mean,
                           'Nec mean': nec_mean,
                           'BK 10x': bk_10x})


# Positive Data
## Filter the columns containing specific text in their names
Compound = [col for col in pos_measure.columns if 'Compound' in col]
dmso_columns = [col for col in pos_measure.columns if 'DMSO' in col]
sts_columns = [col for col in pos_measure.columns if 'STS' in col]
fas_columns = [col for col in pos_measure.columns if 'Fas' in col]
rsl_columns = [col for col in pos_measure.columns if 'RSL' in col]
nec_columns = [col for col in pos_measure.columns if 'Nec' in col]
bk_columns = [col for col in pos_measure.columns if 'BK' in col][0]

## Calculate the mean each group
dmso_mean = pos_measure[dmso_columns].mean(axis=1)
sts_mean = pos_measure[sts_columns].mean(axis=1)
fas_mean = pos_measure[fas_columns].mean(axis=1)
rsl_mean = pos_measure[rsl_columns].mean(axis=1)
nec_mean = pos_measure[nec_columns].mean(axis=1)

# Multiply the BK column by 10
bk_10x = pos_measure[bk_columns] * 10

## Create the new filtered dataframe
pos_filter = pd.DataFrame({'Compound': pos_measure['Compound'],
                           'iden': pos_measure['Identifications'],
                           'DMSO mean': dmso_mean,
                           'STS mean': sts_mean,
                           'Fas mean': fas_mean,
                           'RSL mean': rsl_mean,
                           'Nec mean': nec_mean,
                           'BK 10x': bk_10x})
```

```python
# remove the row with 0 identification
neg_filter = neg_filter[neg_filter['iden'] != 0]
pos_filter = pos_filter[pos_filter['iden'] != 0]

print(len(neg_filter))
print(len(pos_filter))
```

```python
# find the max value in each row then add it to the filter data
# negative data
max_signal = []
for i in list(range(0, len(neg_filter))):
    n = neg_filter.iloc[i, :][2:7].max()
    max_signal.append(n)

neg_filter['Max Signal'] = max_signal

# positive data
max_signal = []
for i in list(range(0, len(pos_filter))):
    n = pos_filter.iloc[i, :][2:7].max()
    max_signal.append(n)

pos_filter['Max Signal'] = max_signal
```

```python
# Check which row has a max signal that is 10x greater than the BK
# negative data
neg_filter['Keep'] = neg_filter['Max Signal'] > neg_filter['BK 10x']

# positive data
pos_filter['Keep'] = pos_filter['Max Signal'] > pos_filter['BK 10x']
```

```python
# Remove the rows that have a max signal that is 10x lesser than the BK
# negative data
neg_filter = neg_filter[neg_filter['Keep'] == True]
neg_filter = neg_filter.drop(['Keep', 'Max Signal'], axis=1)

# positive data
pos_filter = pos_filter[pos_filter['Keep'] == True]
pos_filter = pos_filter.drop(['Keep', 'Max Signal'], axis=1)

print(len(neg_filter))
print(len(pos_filter))
```

```python
# the file path may be different base on the location of the folder
neg_filter.to_csv('/Volumes/Johnathan/1_Metabolomic/1_Data/20231204_Cell-Death_Metabolite_Sup_HMDB/Processed_Data/measurement/neg_filtered.csv')
pos_filter.to_csv('/Volumes/Johnathan/1_Metabolomic/1_Data/20231204_Cell-Death_Metabolite_Sup_HMDB/Processed_Data/measurement/pos_filtered.csv')
```

```python
# Merge the dataframes on the 'Compound' column
## negative mode
neg_2x = pd.merge(neg_filter, neg_sts[['Compound', 'STS 2x']], on = 'Compound', how = 'left')
neg_2x = pd.merge(neg_2x, neg_fas[['Compound', 'Fas 2x']], on = 'Compound', how = 'left')
neg_2x = pd.merge(neg_2x, neg_rsl[['Compound', 'RSL 2x']], on = 'Compound', how = 'left')
neg_2x = pd.merge(neg_2x, neg_nec[['Compound', 'Nec 2x']], on = 'Compound', how = 'left')
### fill the NaN with Faslse
neg_2x.fillna(False, inplace = True)

## positive mode
pos_2x = pd.merge(pos_filter, pos_sts[['Compound', 'STS 2x']], on = 'Compound', how = 'left')
pos_2x = pd.merge(pos_2x, pos_fas[['Compound', 'Fas 2x']], on = 'Compound', how = 'left')
pos_2x = pd.merge(pos_2x, pos_rsl[['Compound', 'RSL 2x']], on = 'Compound', how = 'left')
pos_2x = pd.merge(pos_2x, pos_nec[['Compound', 'Nec 2x']], on = 'Compound', how = 'left')
### fill the NaN with Faslse
pos_2x.fillna(False, inplace = True)
```
