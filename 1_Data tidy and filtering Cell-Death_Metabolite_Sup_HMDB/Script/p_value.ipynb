### Tidy raw data

Rename the column name with row_1

```python
# negative data
neg_measure.columns = list(neg_measure.iloc[1])
neg_measure
```

```python
# positive data
pos_measure.columns = list(pos_measure.iloc[1])
pos_measure
```

Remove row 0 & 1

```python
# negative data
neg_measure = neg_measure.drop([0,1])
neg_measure
```

```python
# positive data
pos_measure = pos_measure.drop([0,1])
pos_measure
```

Extract the necessary columns only

```python
# Negative Data
## Filter the columns containing specific text in their names
dmso_columns = [col for col in neg_measure.columns if 'DMSO' in col]
sts_columns = [col for col in neg_measure.columns if 'STS' in col]
fas_columns = [col for col in neg_measure.columns if 'Fas' in col]
rsl_columns = [col for col in neg_measure.columns if 'RSL' in col]
nec_columns = [col for col in neg_measure.columns if 'Nec' in col]

neg_measure = pd.concat([neg_measure["Compound"], neg_measure[dmso_columns], neg_measure[sts_columns], neg_measure[fas_columns], neg_measure[rsl_columns], neg_measure[nec_columns]], axis=1)
neg_measure
```

```python
# Positive data
## Filter the columns containing specific text in their names
dmso_columns = [col for col in pos_measure.columns if 'DMSO' in col]
sts_columns = [col for col in pos_measure.columns if 'STS' in col]
fas_columns = [col for col in pos_measure.columns if 'Fas' in col]
rsl_columns = [col for col in pos_measure.columns if 'RSL' in col]
nec_columns = [col for col in pos_measure.columns if 'Nec' in col]

pos_measure = pd.concat([pos_measure["Compound"], pos_measure[dmso_columns], pos_measure[sts_columns], pos_measure[fas_columns], pos_measure[rsl_columns], pos_measure[nec_columns]], axis=1)
pos_measure
```

Rename the column

```python
# negative data
neg_measure.columns = ["Compound", 
                       "DMSO-1", "DMSO-2", "DMSO-3",
                       "STS-1", "STS-2", "STS-3",
                       "Fas-1", "Fas-2", "Fas-3",
                       "RSL-1", "RSL-2", "RSL-3",
                       "Nec-1", "Nec-2", "Nec-3"]
neg_measure
```

```python
# positive data
pos_measure.columns = ["Compound",
                          "DMSO-1", "DMSO-2", "DMSO-3",
                          "STS-1", "STS-2", "STS-3",
                          "Fas-1", "Fas-2", "Fas-3",
                          "RSL-1", "RSL-2", "RSL-3",
                          "Nec-1", "Nec-2", "Nec-3"]
pos_measure
```

### Remove the unnecessary row (base on cut data)

```python
# negative data
neg_measure["Compound"].isin(neg_measure_cut["Compound"]).value_counts()
```

```python
neg_measure = neg_measure[neg_measure["Compound"].isin(neg_measure_cut["Compound"])]
neg_measure
```

```python
# positive data
pos_measure["Compound"].isin(pos_measure_cut["Compound"]).value_counts()
```

```python
pos_measure = pos_measure[pos_measure["Compound"].isin(pos_measure_cut["Compound"])]
pos_measure
```

### Merge the data

```python
# negative data
neg_join = pd.merge(neg_measure_cut, neg_measure, on="Compound")
neg_join
```

```python
# positive data
pos_join = pd.merge(pos_measure_cut, pos_measure, on="Compound")
pos_join
```

```python
neg_join.drop(["Unnamed: 0"], axis=1, inplace=True)
neg_join
```

```python
pos_join.drop(["Unnamed: 0"], axis=1, inplace=True)
pos_join
```

```python
neg_join.to_csv("/Volumes/Johnathan/1_Metabolomic/1_Data/20231204_Cell-Death_Metabolite_Sup_HMDB/Processed_Data/cut/neg_measure_cut.csv")
pos_join.to_csv("/Volumes/Johnathan/1_Metabolomic/1_Data/20231204_Cell-Death_Metabolite_Sup_HMDB/Processed_Data/cut/pos_measure_cut.csv")
```
