## This script is to rank the compound based on their fold-change in each treatment

### Import packages

```python
import numpy as np
import pandas as pd
```

### Import data from the csv file

```python
neg_measure = pd.read_csv('/Volumes/Johnathan/1_Metabolomic/1_Data/20231204_Cell-Death_Metabolite_Sup_HMDB/Processed_Data/cut/neg_measure_cut.csv')
pos_measure = pd.read_csv('/Volumes/Johnathan/1_Metabolomic/1_Data/20231204_Cell-Death_Metabolite_Sup_HMDB/Processed_Data/cut/pos_measure_cut.csv')
```

```python
neg_measure
```

```python
pos_measure
```

### Score generate

```python
# negative data
neg_num = len(neg_measure) # number of compounds

total_score = 2 # total sum of score

neg_rank_score = dict(zip(range(1, neg_num + 1), np.linspace(0, 2, num = neg_num)))

print(neg_rank_score)
```

```python
# positive data
pos_num = len(pos_measure) # number of compounds

total_score = 2 # total sum of score

pos_rank_score = dict(zip(range(1, pos_num + 1), np.linspace(0, 2, num = pos_num)))

print(pos_rank_score)
```

### Ranking

```python
# negative data
neg_measure = neg_measure.assign(sts_rank = lambda x: x['sts_fold'].rank(method = 'min', ascending = False),
                                 fas_rank = lambda x: x['fas_fold'].rank(method = 'min', ascending = False),
                                 rsl_rank = lambda x: x['rsl_fold'].rank(method = 'min', ascending = False),
                                 nec_rank = lambda x: x['nec_fold'].rank(method = 'min', ascending = False))
neg_measure
```

```python
# positive data
pos_measure = pos_measure.assign(sts_rank = lambda x: x['sts_fold'].rank(method = 'min', ascending = False),
                                 fas_rank = lambda x: x['fas_fold'].rank(method = 'min', ascending = False),
                                 rsl_rank = lambda x: x['rsl_fold'].rank(method = 'min', ascending = False),
                                 nec_rank = lambda x: x['nec_fold'].rank(method = 'min', ascending = False))
pos_measure
```

### Give the score

```python
# replace the rank with score
for column in neg_measure.columns[11:]:
    neg_measure[column] = neg_measure[column].replace(neg_rank_score)

neg_measure['total_score'] = neg_measure.iloc[:, 11:].sum(axis = 1)
neg_measure.drop(neg_measure.columns[11:15], axis = 1, inplace = True)

neg_measure
```

```python
# positive data
for column in pos_measure.columns[11:]:
    pos_measure[column] = pos_measure[column].replace(pos_rank_score)

pos_measure['total_score'] = pos_measure.iloc[:, 11:].sum(axis = 1)
pos_measure.drop(pos_measure.columns[11:15], axis = 1, inplace = True)

pos_measure
```

### Save to csv file

```python
neg_measure.to_csv('/Volumes/Johnathan/1_Metabolomic/1_Data/20231204_Cell-Death_Metabolite_Sup_HMDB/Processed_Data/human effort/neg_measure_cut_score.csv')
pos_measure.to_csv('/Volumes/Johnathan/1_Metabolomic/1_Data/20231204_Cell-Death_Metabolite_Sup_HMDB/Processed_Data/human effort/pos_measure_cut_score.csv')
```
